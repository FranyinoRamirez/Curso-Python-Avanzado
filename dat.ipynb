```python
def enviar_email(usuario, password, destinatario, mensaje_str):
    """
    Gestinar la conexion con el servidor SMTP y enviar un correo electrónico.
    Retornar true si el correo se envió correctamente, o False en caso contrario.
    """

    servidor_smtp = 'smtp.gmail.com'
    puerto_smtp = 587

    try:
        #Conexion y Seguridad
        servidor = smtplib.SMTP(servidor_smtp, puerto_smtp)
        servidor.starttls()

        #Autenticacion y envio 
        servidor.login(usuario, password)
        servidor.sendmail(usuario, destinatario, mensaje_str)
        print("Correo enviado correctamente.")
        return True
    except Exception as e:
        print(f"Error al enviar el correo: {e}")
        return False
```
